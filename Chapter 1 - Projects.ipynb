```python
# P-1.29
from math import factorial
from random import shuffle
lst = ['c','a','t','d','o','g']
mixes = []
c=0
while c != factorial(len(lst)):
    shuffle(lst) 
    y = lst
    x = ''.join(y)
    if x not in mixes:
        mixes.append(x)
        c+=1
    
print(len(mixes)) # 6!==720
sorted(mixes) # All possible strings (exactly once)
```

```python
# P-1.30
from math import log2

def div_2(n):
    if n <= 2:
        print('invalid entry')
        return
    return log2(n)//1 # The base 2 log, when rounded to the floor, 
                      # can tell you the number of times the integer can
                      # be divide by two before reaching a value < 2.
    
inp = int(input('Enter integer greater than 2: '))
div_2(inp)
```

```python
# P-1.31

def make_change(charge, given, p=[100, 20, 10, 5, 1, 0.25, 0.10, 0.05, 0.01]):
    if given < charge: print('The customer has not given enough!'); return ''
    change = {}
    rem = given-charge
    print('Change back: $'+str(round(rem,2)))
    for i in p:
        num_bills, rem = divmod(rem, i)
        if num_bills: change[i] = int(num_bills)   
    return change

print(make_change(144.51, 150))
```

```python
# P-1.32
def calc():
    choice = 'yes'
    while choice == 'yes':
        inp = int(input())
        inp2 = input()
        assert inp2 in ['+','-','*','/','='], 'Enter operator.'
    
        if inp2=='=':
            print(inp)
            print('Continue? (yes/no): ')
            choice = input()
            if choice == 'yes': continue
            else: break
    
        inp3 = int(input())
    
        if inp2=='+':
            print(inp+inp3)
        elif inp2=='-':
            print(inp-inp3)
        elif inp2=='*':
            print(inp*inp3)
        elif inp2=='/':
            print(inp/inp3)
            
        print('Continue? (yes/no): ')
        choice = input()
calc()
```

```python
# P-1.33
def calc():
    choice = 'yes'
    while choice == 'yes':
        inp = input()
        if inp == 'reset': continue
        inp=float(inp)
            
        inp2 = input()
        assert inp2 in ['+','-','*','/','=','reset'], 'Enter operator.'
        if inp2 == 'reset': continue
        elif inp2=='=':
            print(str(inp)+' = '+str(inp))
            print('Continue? (yes/no): ')
            choice = input()
            if choice == 'yes': continue
            else: break
            
        inp3 = input()
        if inp3 == 'reset': continue
        inp3 = float(inp3)
    
        if inp2=='+':
            print(str(inp)+' + '+str(inp3)+' = '+str(inp+inp3))
        elif inp2=='-':
            print(str(inp)+' - '+str(inp3)+' = '+str(inp-inp3))
        elif inp2=='*':
            print(str(inp)+' * '+str(inp3)+' = '+str(inp*inp3))
        elif inp2=='/':
            print(str(inp)+' / '+str(inp3)+' = '+str(inp/inp3))
            
        print('Continue? (yes/no): ')
        choice = input()
calc()
```

```python
# P-1.35
from math import factorial
n = [x for x in range(5, 101) if x%5==0]
proba = []
for room in n:
    probability = 1 - ( factorial(365)/(365**room * factorial(365-room)) )
    proba.append(probability)
    
for c in range(len(proba)): 
    print('People in room: '+str(n[c])+ '  Probability: '+str(round(proba[c]*100,4))+'%')
```
