```python
# R-3.1
import matplotlib.pyplot as plt
import math

x = [10**i for i in range(0, 4)]

y1 = [8*num for num in x]
y2 = [4*num*math.log(num, 2) for num in x]
y3 = [2*num**2 for num in x]
y4 = [num**3 for num in x]
y5 = [2**num for num in x]

plt.loglog(x, y1, label='8n')
plt.loglog(x, y2, label='4nlogn')
plt.loglog(x, y3, label='2n^2')
plt.loglog(x, y4, label='n^3')
plt.loglog(x, y5, label='2^n')
plt.ylim([10**1, 10**9])
plt.xlim([10**1, 10**3])
plt.legend()
plt.show()
```

```python
# R-3.2
import matplotlib.pyplot as plt
import math

x = [x for x in range(1,101)]

y1 = [8*n*math.log(n, 2) for n in x]
y2 = [2*n**2 for n in x]

plt.plot(x, y1, label='A: 8nlogn')
plt.plot(x, y2, label='B: 2n^2')
plt.xlabel('input size')
plt.ylabel('running time')
plt.xlim([1,1.6])
plt.ylim([0,4])
plt.legend()
plt.show()
# By setting the run times equal to one another, we solve for n≈1.2396277.
# So we see the run-time is better for A until around an input size n of 1.239, thereafter,
# B becomes a better algorithm only after 1.239.
```

```python
# R-3.3
import matplotlib.pyplot as plt
import math

x = [x for x in range(0,101)]

y1 = [40*n**2 for n in x]
y2 = [2*n**3 for n in x]

plt.plot(x, y1, label='A: 40n^2')
plt.plot(x, y2, label='B: 2n^3')
plt.xlabel('input size')
plt.ylabel('running time')
plt.xlim([0,30])
plt.ylim([0,100000])
plt.legend()
plt.show()
# This can be solved easily, again by setting the number of operations
# equal to one another and solving for n (the intersection). Here we see 
# A becomes the faster algorithm for n >= 20.
```

```python
# R-3.7

# The big-Oh notation actually describes a collection of
# functions. 
# Let f(n) = n^3.
# For example, n^2, nlog(n), and n all are 'in'
# O(n^3). 
# The worst case, or largest run-time, would be the quickest
# rising function in O(n^3), which is n^3, as all the other
# functions fall underneath it.

# Therefore, if the the run-time of A is always O(f(n)),
# we also know f(n) is the worst case in the collection.
```

```python
# R-3.8
import matplotlib.pyplot as plt
from math import log

x = list(range(1,61))

y1 = [4*n*log(n, 2)+2*n for n in x]
# 2^10 is a constant and thus left out
y2 = [n for n in x] # 2^log(n) = n for base 2 (we use base 2 logs in CS)
y3 = [3*n + 100*log(n, 2) for n in x]
y4 = [4*n for n in x]
y5 = [2**n for n in x]
y6 = [n**2+10*n for n in x]
y7 = [n**3 for n in x]
y8 = [n*log(n, 2) for n in x]

plt.plot(x, y1, label='4nlogn+2n')
plt.plot(x, y2, label='2^log(n)')
plt.plot(x, y3, label='3n+100log(n)')
plt.plot(x, y4, label='4n')
plt.plot(x, y5, label='2^n')
plt.plot(x, y6, label='n^2+10n')
plt.plot(x, y7, label='n^3')
plt.plot(x, y8, label='nlog(n)')

plt.ylim([0,2000])
plt.legend(loc='upper right')
plt.show()
```

```python
# R-3.23

def example1(S):
    '''Return the sum of the elements in sequence S.'''
    n = len(S) # O(1)
    total = 0 # O(1)
    for j in range(n): # O(n)
        total += S[j] # O(1)+O(1)
    return total # O(1)

# 1+1+n+1+1+1 = n+5 is O(n)
```

```python
# R-3.24

def example2(S):
    '''Return the sum of the elements with even index in sequence S.'''
    n = len(S) # O(1)
    total = 0 # O(1)
    for j in range(0, n, 2): # O(n) -this subtracts 2 for input each time: still linear
        total += S[j] # O(1)+O(1)
    return total # O(1)

# # 1+1+n+1+1+1 = n+5 is O(n)
```

```python
# R-3.25

def example3(S):
    '''Return the sum of the prefix sums of sequence S.'''
    n = len(S) # O(1)
    total = 0 # O(1)
    for j in range(n): # O(n)
        for k in range(1+j): # O(n)*n
            total += S[k] # O(1)+O(1)
    return total # O(1)

# 1+1+n*n+1+1+1 = n^2 + 5 is O(n^2)
```

```python
# R-3.26

def example4(S):
    '''Return the sum of the prefix sums of sequence S.'''
    n = len(S) # O(1)
    prefix = 0 # O(1)
    total = 0 # O(1)
    for j in range(n): # O(n)
        prefix += S[j] # O(1)+O(1)
        total += prefix # O(1)+O(1)
    return total # O(1)

# 1+1+1+n+1+1+1+1+1 = n+8 is O(n)
```

```python
# R-3.27

def example5(A, B):
    '''Return the number of elements in B equal to the sum of prefix sums in A.'''
    n = len(A) # O(1)
    count = 0 # O(1)
    for i in range(n): # O(n)
        total = 0 # O(1)
        for j in range(n): # O(n)*n
            for k in range(1+j): # O(n)*n*n
                total += A[k] # O(1)+O(1)
        if B[i] == total: # O(1)
            count += 1 # O(1)+O(1)
        return count #O(1)
    
# 1+1+n+1+n^2+n^3+1+1+1+1+1+1 = n^3 + n^2 + 9 is O(n^3)
```
