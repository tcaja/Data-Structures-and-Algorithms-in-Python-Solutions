```python
# R-2.4
class Flower:
    """Creates a mutable instance of a certain flower's 
       name, number of petals, and price."""
    
    def __init__(self, name, petals, price): 
        if type(name)   != str:   raise TypeError('name must be a string')
        if type(petals) != int:   raise TypeError('petals must be an int')
        if type(price)  != float: raise TypeError('petals must be a float')
        self._name   = name
        self._petals = petals 
        self._price  = price

    def set_name(self, n):
        if type(n) != str: 
            raise TypeError('name must be a string')
        self._name = n
        
    def get_name(self): return self._name
    
    def set_petals(self, p):
        if type(p) != int: 
            raise TypeError('petals must be an int')
        self._petals = p
        
    def get_petals(self): return self._petals
    
    def set_price(self, pri):
        if type(pri) != float: 
            raise TypeError('price must be a float')
        self._price = pri
            
    def get_price(self): return self._price

x = Flower('Rose', 27, 2.89)

print(x.get_name(), end='\t')
print(x.get_petals(), end='\t')
print(x.get_price())

x.set_name('Tulip')
x.set_petals(16)
x.set_price(0.91)

print(x.get_name(), end='\t')
print(x.get_petals(), end='\t')
print(x.get_price())
```

```python
# R-2.16

# To calculate the number of elements (or length) in
# the Range class, we use the formula 'max(0, (stop-start+step-1)//step)'.

# You can never have a negative length; lowest value for max is then zero.

# Since this works just like the range function, observe:
print(list(range(5,26,5)))
# We know the length is 5. Checking the formula:
# max(0, (26-5+5-1)//5)) == max(0, 5) == 5

# We can check this for negative steps as well:
print(list(range(25,4,-5)))
# Again, the number of elements is 5:
# max(0, (4-25+(-5)-1)//-5) == max(0, -27//-5) == max(0, 5) == 5

# Notice that stop-start will yield a negative numerator in this
# case and the integer division of the negative step returns a
# rounded, positive integer in the max function.
```

```python
# R-2.18
class Progression:
    def __init__(self, start=0):
        self._current = start
    
    def _advance(self):
        self._current += 1
        
    def __next__(self):
        if self._current is None: raise StopIteration()
        else:
            answer = self._current
            self._advance()
            return answer
        
    def __iter__(self):
        return self
    
    def print_progression(self, n):
        print(' '.join(str(next(self)) for j in range(n)))
        
class FibonacciProgression(Progression):
    
    def __init__(self, first=0, second=1):
        super().__init__(first)
        self._prev = second - first
        
    def _advance(self):
        self._prev, self._current = self._current, self._prev + self._current
        
prog = FibonacciProgression(2,2)
x = [next(prog) for i in range(8)]
x[7]
```

```python
# R-2.19
class ArithmeticProgression(Progression):
    def __init__(self, increment=1, start=0):
        super().__init__(start)               
        self._increment = increment

    def _advance(self):                      
        self._current += self._increment
        
# Without straining our poor computers, we know that if
# we increment by 128, then 2**63 // 128 will give us the
# number of increments to reach 2**63:
#  2**63 // 128 == 7.206e16 calls to the next function (a lot).
```

```python
# R-2.22
from abc import ABCMeta, abstractmethod

class Sequence(metaclass=ABCMeta):
    @abstractmethod
    def __len__(self): pass
        
    @abstractmethod
    def __getitem__(self, j): pass
        
    def __contains__(self, val):
        for j in range(len(self)):
            if self[j] == val:
                return True
        return False
    
    def index(self, val):
        for j in range(len(self)):
            if self[j] == val:
                return j
        raise ValueError('value not in sequence')
        
    def count(self, val):
        k = 0
        for j in range(len(self)):
            if self[j] == val:
                k+=1
        return k
    
    def __eq__(self, other):
        if len(self) != len(other): return False
        for j in range(len(self)):
            if self[j] != other[j]:
                return False
        return True
```

```python
# R-2.23
from abc import ABCMeta, abstractmethod
class Sequence(metaclass=ABCMeta):
    @abstractmethod
    def __len__(self): pass
        
    @abstractmethod
    def __getitem__(self, j): pass
        
    def __contains__(self, val):
        for j in range(len(self)):
            if self[j] == val:
                return True
        return False
    
    def index(self, val):
        for j in range(len(self)):
            if self[j] == val:
                return j
        raise ValueError('value not in sequence')
        
    def count(self, val):
        k = 0
        for j in range(len(self)):
            if self[j] == val:
                k+=1
        return k
    
    def __lt__(self, other):
        for j in range(min(len(self), len(other))):
            if self[j] < other[j]:
                return True
        return False
```
