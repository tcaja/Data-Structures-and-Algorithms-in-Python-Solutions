```python
# P-2.33
class Monomial:
    """Class that uses a method to get derivative of a MONOmial."""
    def __init__(self, coef, exp):
        self._coef = coef
        self._exp  = exp
        
    def derivative(self):
        self._coef *= self._exp
        self._exp -= 1
        print(str(self._coef)+'x^'+str(self._exp), end=' ')
    
class Binomial(Monomial):
    """Class that uses a method to get derivative of a BInomial."""
    def __init__(self, coef, exp, co, ex):
        super().__init__(coef, exp)
        self._co = co
        self._ex = ex
        
    def derivative(self):
        self._co *= self._ex
        self._ex -= 1
        super().derivative()
        print('+ '+str(self._co)+'x^'+str(self._ex), end=' ')
        
class Trinomial(Binomial):
    """Class that uses a method to get derivative of a TRInomial."""
    def __init__(self, coef, exp, co, ex, c, e):
        super().__init__(coef, exp, co, ex)
        self._c = c
        self._e = e
        
    def derivative(self):
        self._c *= self._e
        self._e -= 1
        super().derivative()
        print('+ '+str(self._c)+'x^'+str(self._e), end=' ')
```

```python
x = Monomial(5,4)
x.derivative()
```

```python
y = Binomial(5,4, 3,4)
y.derivative()
```

```python
z = Trinomial(5,4, 3,3, 2,4)
z.derivative()
```

```python
# P-2.34
class DocumentLetterCounter:
    """Class reads a document and counts the number of each letter within it,
       and shows a bar plot for letter quantity."""
    
    _alphabet = [chr(x) for x in range(97, 123)]
    
    def __init__(self, file):
        """opens file and assigns it to self._doc"""
        self._doc = open(file).read()
        
    def count_letters(self):
        """Processes and counts number of each letter in document"""
        strng = self._doc.replace('\n', '').lower()
        
        l_counts = {}
        for letter in self._alphabet:
            l_counts[letter] = strng.count(letter)
        
        for k, v in l_counts.items():
            print(k+'| '+'X'*v+'('+str(v)+')')
              
road = DocumentLetterCounter('this.txt')
road.count_letters()
```

```python
# P-2.35
import random as r

class Alice:
    
    def __init__(self):
        self._packet = []
        
    def create_packet(self):
        packet = []
        for i in range(r.randint(0, 5)):
            packet.append(r.randint(0, 9))
        self._packet = packet
        
    def get_packet(self):
        return self._packet
    
class Internet(Alice):
    
    def __init__(self): pass
        
    def process_mail(self, packet): 
        if len(packet):
            print('Mail!', end='\t')
            return True
        else:
            print('No mail!')
            return False
        
        
class Bob(Internet):
    
    def __init__(self):
        self._box = []
        
    def check_mail(self, mail):
        if super().process_mail(mail):
            self._box = mail
            print(self._box)
            
for i in range(24):
    print('Hour '+str(i+1)+': ', end='')
    x = Alice()
    x.create_packet()
    y = Bob()
    y.check_mail(x.get_packet())
```

```python
# P-2.37
import random as r
class Ecosystem:
    
    def __init__(self): self.river = [r.choice([Bear(), Fish(), None]) for i in range(6)]
    
    def show_river(self): 
        for item in self.river: 
            if item != None: 
                print(str(item)+' '+str(item.gender), end=' | ')
            else: print(str(item), end=' | ')
        print('\n')
        
    def move(self, ind1, ind2):
        self.river[ind2], self.river[ind1] = self.river[ind1], self.river[ind2]
        
    def reproduce(self, ind):
        try:
            j = self.river.index(None)
            self.river[j] = self.river[ind]
        except ValueError:
            raise ValueError('Environment is full!')
        
    def die(self, ind):
        self.river[ind] = None
    
        
class Animal(Ecosystem):
    def __init__(self): pass
    
        
class Bear(Animal):
    
    def __init__(self): 
        self.gender = r.randint(0,1)
        self.strength = r.randint(1,5)
    
    def __str__(self): return 'Bear'
        
class Fish(Animal):
    
    def __init__(self): 
        self.gender = r.randint(0,1)
        self.strength = r.randint(1,5)
    
    def __str__(self): return 'Fish'
    
eco = Ecosystem()
eco.show_river() # 0 or 1 displays gender

for i in range(len(eco.river)-1):
    
    if eco.river[i] == None: continue
        
    elif str(eco.river[i]) == 'Bear' and str(eco.river[i+1]) == 'Fish':
        eco.die(i+1)
    elif str(eco.river[i]) == 'Fish' and str(eco.river[i+1]) == 'Bear':
        eco.die(i)
        
    elif str(eco.river[i]) == str(eco.river[i+1]):
        
        if eco.river[i].gender != eco.river[i+1].gender:
            eco.reproduce(i)
        else:
            if eco.river[i].strength > eco.river[i+1].strength:
                eco.die(i+1)
            else: eco.die(i)
                
    else: eco.move(i, i+1)
        
    eco.show_river()
```

```python
# P-3.
class eBook:
    
    def __init__(self):
        self._library = []
        self._store = ['Alice in Wonderland.txt', 'A Seperate Reality.txt', 'Moby Dick.txt']
        
    def view_bookstore(self):
        print('Book Store:')
        for index, item in enumerate(self._store):
            print(str(index+1)+'. '+str(item).replace('.txt', ''))
            
    def view_library(self):
        print('\nYour library:')
        for index, item in enumerate(self._library):
            print(str(index+1)+'. '+str(item).replace('.txt', ''))
            
    def purchase_book(self, val):
        self.view_bookstore()
        if val not in [1,2,3]:
            print('\ninvalid book number')
        else:
            self._library.append(self._store[val-1])
            print('\nYou purchased '+str(self._store[val-1]).replace('.txt', '') \
                 +', thank you!')
            
    def read_book(self, val):
        
        self.view_library()
        
        with open(self._library[val-1], 'r') as file:
            print('\nOpening "'+self._library[val-1]+'"...\n')
            f = file.read()
            print(f)
            
x = eBook()
x.purchase_book(2)
x.read_book(1)
```

```python
from abc import ABCMeta, abstractmethod
import math

class Polygon(metaclass=ABCMeta):
    '''Abstract base class for various polygons'''
    @abstractmethod
    def area(): pass
        
    @abstractmethod
    def perimeter(): pass
    
class Octagon(Polygon):
    '''Creates Octagon object with side length of parameter given'''
    
    def __init__(self, side_length):
        self._slength = side_length
        
    def area(self): 
        return 2*(self._slength**2)*(1+2**0.5)
    
    def perimeter(self): 
        return 8*self._slength
    
class Hexagon(Polygon):
    '''Creates Hexagon object with side length of parameter given'''
    
    def __init__(self, side_length):
        self._slength = side_length
        
    def area(self): 
        return (3*3**0.5)/2 * self._slength**2
    
    def perimeter(self): 
        return 6*self._slength
    
class Pentagon(Polygon):
    '''Creates Pentagon object with side length of first parameter
       given and an apothem from the second parameter'''
    
    def __init__(self, side_length, apothem):
        self._slength = side_length
        self._apothem = apothem
        
    def area(self): 
        return (5/2)*self._slength*self._apothem
    
    def perimeter(self): 
        return 5*self._slength
    
class Square(Polygon):
    '''Creates Square object with side lengths of first parameter'''
    
    def __init__(self, side_length):
        self._slength = side_length
        
    def area(self): 
        return self._slength**2
    
    def perimeter(self): 
        return 4*self._slength
    

class Rectangle(Polygon):
    '''Creates Rectangle object with side lengths of first parameter
       and side widths as second'''
    
    def __init__(self, length, width):
        self._length = length
        self._width  = width
        
    def area(self): 
        return self._length*self._width
    
    def perimeter(self): 
        return 2*self._length+2*self._width
    
class Triangle(Polygon):
    pass
 
class EquilateralTriangle(Triangle):
    '''Create EquilateralTriangle object with side lengths of first parameter'''
    
    def __init__(self, length):
        self._slength = length
        
    def area(self): 
        return (3**0.5/4)*self._slength**2
    
    def perimeter(self): 
        return 3*self._slength
    
class IsoscelesTriangle(Triangle):
    '''Create IsoscelesTriangle object with base for first parameter
       and height for second'''
    
    def __init__(self, base, height):
        self._base = base
        self._height = height
        
    def area(self): 
        return 0.5*self._base*self._height
    
    def perimeter(self): 
        return('No parameter for this object')
    
    
polylist = ['octagon', 'hexagon', 'pentagon', 'square', 'rectangle', 'triangle']

inp = input('Enter a polygon to make!: ')

if inp.lower() not in polylist:
    print('We are unable to create that request')
    
elif inp.lower() == 'octagon':
    num = float(input('Enter a length for the sides of your octagon: '))
    poly = Octagon(num)
    print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    
elif inp.lower() == 'hexagon':
    num = float(input('Enter a length for the sides of your hexagon: '))
    poly = Hexagon(num)
    print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    
elif inp.lower() == 'pentagon':
    num = float(input('Enter a length for the sides of your pentagon: '))
    num2 = float(input('Enter apothem (length from bottom to center) for pentagon: '))
    poly = Pentagon(num, num2)
    print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    
elif inp.lower() == 'square':
    num = float(input('Enter a length for the sides of your square: '))
    poly = Square(num)
    print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    
elif inp.lower() == 'rectangle':
    num = float(input('Enter a length for your rectangle: '))
    num2 = float(input('Enter a width for your rectangle: '))
    poly = Rectangle(num, num2)
    print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    
elif inp.lower() == 'triangle':
    choice = input('Equilateral or isosceles?: ')
    if choice.lower() == 'equilateral':
        num = float(input('Enter a length for the sides of your triangle: '))
        poly = EquilateralTriangle(num)
        print('\nArea: '+str(poly.area())+'\nPerimeter: '+str(poly.perimeter()))
    elif choice.lower() == 'isosceles':
        num = float(input('Enter your length for 2 of the sides: '))
        num2 = float(input('Enter length for base side: '))
        poly = IsoscelesTriangle(num, num2)
        print('\nArea: '+str(poly.area()))
    else: print('invalid entry')
```
